```python
import fuzzywuzzy
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
n=df17.shape[0]
final=pd.DataFrame(columns=["name17","id17","name","score"])

for i in range(0,n):
    input_name=df17.iloc[i,2]
    match=process.extractOne(input_name,df_shadow.iloc[:,3],scorer=fuzz.ratio)[0]
    name=df_shadow[df_shadow["name_clean"]==match].iloc[0,0]
    score=process.extractOne(input_name,df_shadow.iloc[:,3],scorer=fuzz.ratio)[1]
    final.loc[i]=[df17.iloc[i,0],df17.iloc[i,1],name,score]
    print(i," ",df17.iloc[i,1],"  ",name,score)
```

```python
final.to_excel("final.xlsx")
```
