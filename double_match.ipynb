```python
df17=pd.read_excel("2017new.xlsx")
df17["name_clean"]=[i.upper().replace("NATIONAL ASSOCIATION","").replace("NATIONAL BANK","NB").replace("BANK","BK").replace(" NA","").replace("COMMUNITY","CMNTY").replace("FEDERAL CREDIT UNION","FCU").replace("CREDIT UNION","CU").replace("1ST","FIRST").replace(" CORPORATION","").replace(" TRUST","").replace(" COMPANY","").replace(" CO","").replace(" AND","&").replace(".","").replace(",","").replace("'","").replace(" THE","").replace("LLC","").replace("INC","") for i in list(df17.iloc[:,1])]
df17.head()
```

```python
def first(words):
    split_word = words.split(" ")
    for i in split_word: 
        if len(i) > 3:
            return i
    return split_word[-1]
```

```python
def second(words):
    cnt=0
    temp=[]
    split_word=words.split(" ")
    while cnt<=2:
        for i in split_word:
            if len(i)>2:
                temp.append(i)
                cnt=cnt+1
                
    return " ".join(temp[0:2])

print(second("BANK OF AMERICA"))
```

```python
import fuzzywuzzy
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
leftover_first=[first(i) for i in leftover["name_clean"]]
new_first=[first(i) for i in df17["name_clean"]]
#dict1=dict(zip(leftover["name"],leftover_first))
#dict2=dict(zip(df17["name"],new_first))

lmatch = pd.DataFrame(columns=["respondent_name", "leftover", "score"])
N = 0

for i in leftover_first:
    match, score =process.extractOne(i, new_first)
    lmatch.loc[N] = [leftover.iloc[N,0],df17.iloc[new_first.index(match),1],score]
    N += 1
    print(leftover.iloc[N,0],df17.iloc[new_first.index(match),1],score)
```

```python
lmatch.to_excel("lmatch.xlsx")
```
