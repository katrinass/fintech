```python
def first(words):
    split_word = words.split(" ")
    for i in split_word: 
        if len(i) > 3:
            return i
    return split_word[0]
```

```python
panel18=pd.read_excel("2018panel.xlsx")  
panel18.loc[0]["respondent_name"] = "XXXXxXX  XXXXX   XSSDASDSAD  WADSAD"
panel18["name_clean"]=[i.upper().replace("NATIONAL ASSOCIATION","").replace(" & "," AND ").replace("&"," AND ").replace(",","").replace(".","").replace(" INC","").replace(" LLC","").replace(" THE","").replace(" COMPANY","").replace(" CORPORATION","").replace(" CORP","").replace(" CO","").replace(" NA","").replace("TRUST","") for i in list(panel18.iloc[:,1])]
panel18["new_first"]=[first(i) for i in panel18["name_clean"]]
panel18.head()
```

```python
new_first=[]
for i in panel18["name_clean"]:
    split_word = i.split(" ")
    for j in range(len(split_word)):
        if j == len(split_word) - 1:
            new_first.append(split_word[j]); break
        elif len(split_word[j]) > 3:
            new_first.append(split_word[j] + ' ' + split_word[j + 1]); break 
```

```python
leftover_first = []
for i in leftover["name_clean"]:
    split_word = i.split(" ")
    for j in range(len(split_word)):
        if j == len(split_word) - 1:
            leftover_first.append(split_word[j]); break
        elif len(split_word[j]) > 3:
            leftover_first.append(split_word[j] + ' ' + split_word[j + 1]); break    
```

```python
lmatch = pd.DataFrame(columns=["index", "name18", "leftover", "score"])
N = 0

for i in leftover_first:
    match, score = process.extractOne(i, new_first)
    lmatch.loc[N] = [panel18.iloc[new_first.index(match),0], \
                     leftover.iloc[N,0],\
                     panel18.iloc[new_first.index(match),1],\
                     score]
    N += 1
    print(lmatch.loc[N - 1].tolist())
```
