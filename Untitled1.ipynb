```python
def first(words):
    split_word = words.split(" ")
    for i in split_word: 
        if len(i) > 3:
            return i
    return split_word[-1]
```

```python
def second(words):
    cnt=0
    temp=[]
    split_word=words.split(" ")
    while cnt<=2:
        for i in split_word:
            if len(i)>2:
                temp.append(i)
                cnt=cnt+1
                
    return " ".join(temp[0:2])

print(second("BANK OF AMERICA"))
```

```python
import numpy as np
def second(word):
    split_word = np.array(word.split(" "))
    selected = split_word[[len(i) > 2 for i in split_word]]
    return " ".join(selected[0:2])
```

```python
shadow_two=[second(i) for i in df_shadow["name_clean"]]
new_two=[second(i) for i in df17["name_clean"]]

t2match = pd.DataFrame(columns=["respondent_name", "leftover", "score"])

N = 0

for i in new_two:
    match, score =process.extractOne(i, shadow_two)
    t2match.loc[N] = [df17.iloc[N,1],df_shadow.iloc[shadow_two.index(match),0],score]
    N += 1
    print(N,"  ",df17.iloc[N,1],"  ",df_shadow.iloc[shadow_two.index(match),0],score)
```
